#### Image Translation

```python
import cv2
import numpy as np

image = cv2.imread('dogcat.png')

height, width = image.shape[:2]

quarter_height, quarter_width = height / 4, width / 4

T = np.float32([[1, 0, quarter_width], [0, 1, quarter_height]])

img_translation = cv2.warpAffine(image, T, (width, height))

cv2.imshow('Translation', img_translation)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

### 2. Image Processing and Enhancement

#### Blurring an Image
Là một kỹ thuật để giảm độ sắc nét và chi tiết, làm cho ảnh có vẻ mượt mà hơn. Nó hoạt động bằng cách tính trung bình các giá trị pixel xung quanh mỗi pixel. Nó hữu ích trong các tình huống cần giảm thiểu nhiễu hoặc giảm độ sắc nét, chuẩn bị hình ảnh cho các mô hình thị giác máy tính

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("dogcat.png")
height, width = img.shape[:2]
resize_img = cv2.resize(img, (300, 400))
original_image = cv2.cvtColor(resize_img, cv2.COLOR_RGB2BGR)

#1. Gaussian Blur
Gaussian = cv2.GaussianBlur(resize_img, (15, 15), 0)
Gaussian_rgb = cv2.cvtColor(Gaussian, cv2.COLOR_RGB2BGR)
#2. Median Blur
median = cv2.medianBlur(resize_img, 11)
median_rgb = cv2.cvtColor(median, cv2.COLOR_RGB2BGR)
#3. Bilateral Blur
bilateral = cv2.bilateralFilter(resize_img, 15, 150, 150)
bilateral_rgb = cv2.cvtColor(bilateral, cv2.COLOR_RGB2BGR)
titles = ['Original Image', 'Gaussian Blurred Image', 'Median Blur', 'Bilateral Blur']
images = [original_image, Gaussian_rgb, median_rgb, bilateral_rgb]
for i in range(4):
    plt.subplot(2, 2, i + 1)
    plt.title(titles[i])
    plt.axis('off')
    plt.imshow(images[i])
```

### 3. Image segmentation and thresholding
Kỹ thuật giúp chia một hình ảnh thành các vùng hoặc các đối tượng có ý nghĩa

#### Simple Thresholding


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('dogcat.png')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#Binary Threshold
_, thresh_binary = cv2.threshold(gray_image, 120, 255, cv2.THRESH_BINARY)
#Binary Threshoad Inverted
_, thresh_binary_inv = cv2.threshold(
    gray_image, 120, 255, cv2.THRESH_BINARY_INV)
#Truncated Threshold
_, thresh_trunc = cv2.threshold(gray_image, 120, 255, cv2.THRESH_TRUNC)
#To Zero Threshold
_, thresh_tozero = cv2.threshold(gray_image, 120, 255, cv2.THRESH_TOZERO)
#To Zero Inverted Threshold
_, thresh_tozero_inv = cv2.threshold(gray_image, 120, 255, cv2.THRESH_TOZERO_INV)

images = [gray_image, thresh_binary, thresh_binary_inv, thresh_trunc, thresh_tozero, thresh_tozero_inv]
types = ['Original', 'Binary Threshold', 'Binary Threshoad Inverted', 'Truncated Threshold', 'To Zero Threshold', 'To Zero Inverted Threshold']

for i in range(6):
    plt.subplot(2, 3, i + 1)
    plt.title(types[i])
    plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_RGB2BGR))
    plt.axis('off')
```

### 4. Feature Detection

#### Find Co-ordinates or Contours

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

font = cv2.FONT_HERSHEY_COMPLEX
#load image
img2 = cv2.imread('shape.png', cv2.IMREAD_COLOR)
img = cv2.imread('shape.png', cv2.IMREAD_GRAYSCALE)
#binariza image
_, threshold = cv2.threshold(img, 110, 255, cv2.THRESH_BINARY)
#find contours
contours, _ = cv2.findContours(threshold, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

for cnt in contours:
    #approximate and draw contour
    approx = cv2.approxPolyDP(cnt, 0.009 * cv2.arcLength(cnt, True), True)
    cv2.drawContours(img2, [approx], 0, (0, 0, 255), 5)

    #flatten points
    n = approx.ravel()
    i = 0
    for j in n:
        if i % 2 == 0: #x, y coords
            x, y = n[i], n[i + 1]
            coord = f"{x} {y}"
            if i == 0: #first point
                cv2.putText(img2, "Arrow tip", (x, y), font, 0.5, (255, 0, 0))
            else:
                cv2.putText(img2, coord, (x, y), font, 0.5, (0, 255, 0))
        i += 1

#show result
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.axis("off")
```

#### Loading Video

```python
import cv2
cap = cv2.VideoCapture("video.mp4")
if not cap.isOpened():
    exit()
while True:
    ret, frame = cap.read()

    if not ret:
        break
    cv2.imshow("Video", frame)
    #Press q to quit
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
def main():

    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        print("Error: Could not open webcam.")
        return

    while True:
        ret, frame = cap.read()
        if not ret:
            print("Error: Could not read frame.")
            break

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

        for (x, y, w, h) in faces:
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

        cv2.imshow('Face Detection', frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
